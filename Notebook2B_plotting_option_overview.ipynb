# Option 1: Matplotlib, or most commonly matplotlib.pyplot 

link to online resources:
https://matplotlib.org/

````python
#import the matplotlib python library
import matplotlib.pyplot as plt
````

````python
#and make the plots appear in the notebook
%matplotlib inline 
````

````python
# numeric python library
import numpy as np
````

````python
import matplotlib.style as mplstyle
mplstyle.available
````

````python
mplstyle.use('classic')
````

````python
# we can generate so basic values for x and y to plot

x = np.random.randint(0,100,size=500)
y = np.power(x+3,2)

#then we can plot these on a figure
fig = plt.scatter(x,y)
fig2 = plt.scatter(x,x)

# can add and control many details of the figure
plt.title('An example of a figure')
plt.figure(figsize=(16,12))
plt.figaspect(4)

# and then have a look
plt.show()
````

````python
fig, axs = plt.subplots(1,2)
axs[0].scatter(x,y)
axs[1].bar(x,y)

plt.show()
````

# Option 3: Plotting within Pandas and other Python Libraries

````python
# For handling dataFrames and manipulating data
import pandas as pd
````

````python
#define your data as a tabular DataFrame

df =pd.DataFrame(data={'x':x, 'y':y})
df.head()
````

# Option 4: Plotting with Bokeh and Holoviews, via HVPlot

````python
#%conda install hvplot
````

````python
import hvplot.pandas
````

````python
# you can plot from the DataFrame just by calling hvplot()

df.hvplot.scatter('x', 'y',c='y', cmap='fire').opts(width=800, title='Using hvplot', padding=0.1)
````

````python
#make another column based on a cumilative sum of 'x'

df['cumsum_x']= df.sort_values('x')['x'].cumsum()
````

````python
# name the plots as you combine them with an '*' symbol to overlay 

doubleplot =(df.hvplot.scatter('x', 'y',c='y', cmap='fire').opts(width=800, title='Using hvplot', padding=0.1)*
 df.hvplot.scatter('x','cumsum_x',c='y', cmap='Blues', alpha=0.2))

#the define more options for displaying this plot

doubleplot.opts(width=800,height=600, title='Using hvplot', padding=0.1)
````
